```python
#############################
# Pima indian   
##############################

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers # lr  
import numpy as np
import tensorflow as tf
import pandas as pd

data = pd.read_csv(r"pima-indians-diabetes.csv", dtype=np.float32)
dataset = np.array(data)
#data = np.loadtxt(r"pima-indians-diabetes.csv", delimiter=',', dtype=np.float32)
print(dataset)
x_train = dataset[:700,:-1]
y_train = dataset[:700:,[-1]]

x_test = dataset[700:,:-1]
y_test = dataset[700::,[-1]]

model = Sequential()
model.add(Dense(200, input_dim=8, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
#    
model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=0.001), metrics=['accuracy'])
#  
model.fit(x_train,y_train, epochs=1000, batch_size=10)
# 
score = model.evaluate(x_test,y_test)
print("%s : %.4f%%" %(model.metrics_names[1], score[1]*100))


```
