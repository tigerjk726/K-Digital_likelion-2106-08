```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras import optimizers # lr 조정할때 쓰임
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
np.random.seed(0)
tf.random.set_seed(0)

##############################
# 이항분류로 binary사용, Accuracy 구하기
################################

df = pd.read_csv('mushrooms.csv', names = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 
                                           'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 
                                           'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 
                                           'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color',
                                           'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'])

df.describe()
```

```python
e = LabelEncoder()
df=df.drop(index=0)

for i in df.columns:
    df[i] = e.fit_transform(df[i])

df.head()
```

```python
dataset = df.values
x = dataset[:, 1:23]
y = dataset[:, 0]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
print("x_train.shape =", x_train.shape)
print("x_test.shape =", x_test.shape)
print("y_train.shape =", y_train.shape)
print("y_test.shape =", y_test.shape)
print(y_test[0:5])
      
model = Sequential()
model.add(Dense(50, input_dim=22, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델 학습과정 설정 
model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=0.01), metrics=['accuracy'])

# 모델 학습시키기
model.fit(x,y, epochs=20, batch_size=10)
model.summary()
#모델 평가하기
score = model.evaluate(x_test,y_test)

print("%s : %.4f%%" %(model.metrics_names[1], score[1]*100))
```

```python
##############################
# 다항분류로 categoical 사용, Accuracy 구하기
################################


df = pd.read_csv('mushrooms.csv', names = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 
                                           'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 
                                           'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 
                                           'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color',
                                           'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'])

e = LabelEncoder()

df=df.drop(index=0)

df.describe()

for i in df.columns:
    df[i] = e.fit_transform(df[i])
    
dataset = df.values

x = dataset[:,1:]
y = dataset[:,0]

print(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
print("x_train.shape =", x_train.shape)
print("x_test.shape =", x_test.shape)
print("y_train.shape =", y_train.shape)
print("y_test.shape =", y_test.shape)

#Binary를 Multi_class로 
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)

# categoical로 이항을 다항으로 변환 후
print("categoical로 이항을 다항으로 변환 후")
print("y_train.shape =", y_train.shape)
print("y_test.shape =", y_test.shape)
print(y_test[0:5])

model = Sequential()
model.add(Dense(50, input_dim=22, activation='relu'))
model.add(Dense(2, activation='softmax'))

# 모델 학습과정 설정 
model.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(lr=0.01), metrics=['accuracy'])

# 모델 학습시키기
model.fit(x_train, y_train, epochs=50, batch_size=10)

#모델 평가하기
score = model.evaluate(x_train, y_train)

print("%s : %.4f%%" %(model.metrics_names[1], score[1]*100))
```
