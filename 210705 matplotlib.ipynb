```python
import numpy as np
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("x축과 y축의 tick label 설정")
plt.plot(x, c)
plt.xticks([-np.pi, np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1,0,+1])
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("문자열 정의")
plt.plot(x, c)
plt.xticks([-np.pi, np.pi / 2, 0, np.pi / 2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$',r'$+\pi$'])
plt.yticks([-1,0,+1],["Low", "Zero", "High"])
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.title("파일제목")
plt.plot(x,c, ls="--", label="cosine")
plt.plot(x,s, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.plot(x,c,label = "cosine")
plt.title("legend")
plt.xlabel("eee")
plt.ylabel("ddd")
plt.legend(loc=1)
plt.show()
```

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1) # 비율 2 바이 1 로 하되 1번째 한다.
plt.plot(x1,y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2) # 비율 2 바이 1로 하되 2번째 한다.
plt.plot(x2,y2, 'r.-')
plt.title('time (s)')
plt.ylabel('Undamped')
print(ax2)
plt.tight_layout()
plt.show()
```
