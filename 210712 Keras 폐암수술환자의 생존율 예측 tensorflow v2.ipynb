```python
# 폐암 수술환자의 생존율 예측하기
#딥러닝을 구동하는데 필요한 케라스 함수를 불러옴
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#필요한 라이브러리를 불러옵니다. 
import numpy as np
import tensorflow as tf
import pandas as pd
#실행할때마다 같은 결과를 출력하기 위해 설정하는 부분
np.random.seed(3)
tf.random.set_seed(3)
#준비된 수술환자 데이터를 불러들입니다.
Data = pd.read_csv(r"ThoraricSurgery.csv")
Data1 = np.array(Data)
#환자의 기록과 수술결과를 X와 Y로 구분하여 저장
X = Data1[:,0:-1]
Y = Data1[:,[-1]]
```

```python
#딥러닝 구조를 결정합니다.(모델을 설정하고 실행하는 부분)
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#딥러닝을 실행
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X,Y, epochs=100, batch_size=10)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf
import pandas as pd

np.random.seed(0)
tf.random.set_seed(0)

Data = np.loadtxt("ThoraricSurgery.csv", delimiter=",")
X = Data[:,0:-1]
Y = Data[:,[-1]]

model = Sequential()
model.add(Dense(30, input_dim=17, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X,Y, epochs=100, batch_size=10)
print("\nAccuracy : %.4f" %(model.evaluate(X,Y)[1]))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers # lr 조정할때 쓰임

import numpy as np
import tensorflow as tf
import pandas as pd

np.random.seed(0)
tf.random.set_seed(0)

X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])

model = Sequential()
model.add(Dense(4, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=0.01), metrics=['accuracy']) 
# learningrate는 optimzier의 adam에 들어가있음
# opt = optimizers.Adam(lr=0.1)

model.fit(X,Y, epochs=50, batch_size=10)

print("\nAccuracy : %.4f" %(model.evaluate(X,Y)[1]))
```

```python
#############################
# 환자의 데이터에 따른 사망관련, x값에 대한 y의 시그모이드회귀함수 keras로 구현
##############################

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers # lr 조정할때 쓰임
import numpy as np
import tensorflow as tf

data = np.loadtxt("data-03-diabetes.csv", delimiter=",")
X = data[:,:-1]
Y = data[:,[-1]]

model = Sequential()
model.add(Dense(1000, input_dim=8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=0.01), metrics=['accuracy'])
model.fit(X,Y, epochs=350, batch_size=10)

print("\nAccuracy : %.4f" %(model.evaluate(X,Y)[1]))
```

```python
#############################
# 평균기온에 따른 열대야 발생현황, x값에 대한 y의 시그모이드회귀함수 keras로 구현
##############################

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers # lr 조정할때 쓰임
import numpy as np
import tensorflow as tf
import pandas as pd

data = pd.read_csv(r"jeju.csv")
data1 = np.array(data)
X = data1[1:,:-1]
Y = data1[1:,[-1]]

model = Sequential()
model.add(Dense(100, input_dim=5, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(lr=0.01), metrics=['accuracy'])
hist = model.fit(X,Y, epochs=100, batch_size=10)

print("\nAccuracy : %.4f" %(model.evaluate(X,Y)[1]))

acc = hist.history['acc']
val_acc = history.history['val_acc']   # 테스트셋 오차
loss = hist.history['loss']
val_loss = hist.history['val_loss']

epochs = range(1, len(acc) + 1)


plt.plot(epochs, loss, 'bo', label = 'Trainset_loss')
plt.plot(epochs, val_loss, 'b', label = 'Validation loss')
plt.grid()  
plt.xlabel('epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
