```python
import numpy as np
import cv2

switch_case = {
    ord('a') : 'a키 입력', # ord()함수 - 문자를 아스키코드로 변환
    ord('b') : 'b키 입력',
    0X41 : 'A키 입력',
    int('0X42', 16) : 'B키 입력', # 16진수인 0*42를 10진수로
    2424832 : '왼쪽 화살표키 입력',
    2490368 : '윗쪽 화살표키 입력',
    2555904 : '오른쪽 화살표키 입력',
    2621440 : '아래쪽 화살표키 입력'
}
image = np.ones((200,300), np.uint8)
cv2.namedWindow('Keyboard Event')
cv2.imshow('Keyboard Event', image)
while True:
    key = cv2.waitKeyEx(100)
    if key == 27 :
        break
    try : 
        result = switch_case[key]
        print(result)
    except KeyError:
        result = -1
        
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print("마우스 왼쪽 버튼")
    if event == cv2.EVENT_RBUTTONDOWN:
        print("마우스 오른쪽 버튼")
    if event == cv2.EVENT_RBUTTONUP:
        print("마우스 오른쪽 버튼 떼기")
    if event == cv2.EVENT_LBUTTONDBLCLK:
        print("마우스 왼쪽 버튼 더블클릭")
        
image = np.full((200,300), 255, np.uint8)
title1, title2 = 'Mouse Event1', 'Mouse Event2'
cv2.imshow(title1, image)
cv2.imshow(title2, image)

cv2.setMouseCallback('Mouse Event1', onMouse)
cv2.waitKey(0)        
cv2.destroyAllWindows()
```

```python
###############################
#  원 그리기

import numpy as np
import cv2

orange, blue, cyan = (0, 165, 255), (255,0,0),(255,255,0)
white, black = (255, 255, 255), (0,0,0)
image = np.full((300,500,3), white, np.uint8)

center = (image.shape[1]//2, image.shape[0]//2)
pt1, pt2 = (300,50), (100,220)
shade = (pt2[0] + 2, pt2[1] + 2)

cv2.circle(image, center, 100, blue)
cv2.circle(image, pt1, 50, orange, 2)
cv2.circle(image, pt2, 70, cyan, -1)

font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(image, 'center_blue', center, font, 1.0, blue)
cv2.putText(image, 'pt1_orange', pt1, font, 1.0, orange)
cv2.putText(image, 'pt2_cyan', shade, font, 1.2, black, 2)
cv2.putText(image, 'pt2_cyan', pt2, font, 1.2, cyan, 1)

title = 'Draw circles'
cv2.namedWindow(title)
cv2.imshow(title, image)
cv2.waitKey(0)        
cv2.destroyAllWindows()
```

```python
###############################
# 원과 랜덤값10개의 사각형 그리기

import numpy as np
import cv2
import math
import random

blue, green, red = (255, 0, 0), (0,255,0),(0,0,255)

image = np.full((400,400,3), white, np.uint8)

center = (image.shape[1]//2, image.shape[0]//2)

cv2.circle(image, center, 100, blue, cv2.FILLED)

plus = []
for i in range(10):
    x = random.randint(0,400)
    y = random.randint(0,400)
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(image, '+', (x,y), font, 1.0, red)
    c = math.sqrt((x-image.shape[1]//2)**2 + (y-image.shape[1]//2)**2)
    plus.append([c,(x,y)])

xx = sorted(plus)
        
cv2.line(image, xx[-1][1], xx[-2][1], green)
cv2.line(image, xx[-2][1], xx[-3][1], green)
cv2.line(image, xx[-3][1], xx[-4][1], green)
cv2.line(image, xx[-4][1], xx[-1][1], green)

title = 'Drawing'
cv2.namedWindow(title)
cv2.imshow(title, image)
cv2.waitKey(0)        
cv2.destroyAllWindows()
```

```python
###############################
# 원과 랜덤값10개의 사각형 그리기

import numpy as np
import cv2

def onMouse(event, x, y, flags, param):
    global title, pt
    if event == cv2.EVENT_LBUTTONDOWN:
        if pt[0] <= 0:
            pt = (x, y)    # 시작좌표 지정
        else :
            cv2.rectangle(image, pt, (x, y), (255,0,0), 2)
            cv2.imshow(title, image)
            pt = (-1, -1)   # 시작좌표 초기화
    elif event == cv2.EVENT_RBUTTONDOWN:
        if pt[0] <= 0:
            pt = (x, y)    # 시작좌표 지정
        else :
            dx, dy = pt[0] - x, pt[1] - y
            radius = int(np.sqrt(dx*dx + dy*dy))
            cv2.circle(image, pt, radius, (0,0,255), 2)
            cv2.imshow(title, image)
            pt = (0, 0)   # 시작좌표 초기화
    
image = np.full((400,400,3), white, np.uint8)
pt = (-1, -1)

title = 'Drawing'
cv2.namedWindow(title)
cv2.imshow(title, image)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)        
cv2.destroyAllWindows()
```

```python
###############################
# 영상 파일 읽기

import numpy as np
import cv2

def print_matInfo(name, image):
    if image.dtype == 'uint8': mat_type = 'CV_8U'
    elif image.dtype == 'int8': mat_type = 'CV_8S'
    elif image.dtype == 'uint16': mat_type = 'CV_16U'
    elif image.dtype == 'int16': mat_type = 'CV_16S'
    elif image.dtype == 'float32': mat_type = 'CV_32F'
    elif image.dtype == 'float65': mat_type = 'CV_64F'
    nchannel = 3 if image.ndim == 3 else 1
    
    print('%12s: depth(%s), channels(%s) -> mat_type(%sC%d)'
         % (name,image.dtype, nchannel, mat_type, nchannel))

title1, title2 = 'gray2gray','gray2color'
gray2gray = cv2.imread('images/read_gray.jpg', cv2.IMREAD_GRAYSCALE)
gray2color = cv2.imread('images/read_gray.jpg', cv2.IMREAD_COLOR)
if (gray2gray is None or gray2color is None):
    raise Exception('영상읽기에러')
    
print('행렬 좌표(100,100) 최소값')
print('%s %s' %(title1, gray2gray[100,100]))
print('%s %s\n' %(title2, gray2color[100,100]))
print_matInfo(title1, gray2gray)
print_matInfo(title2, gray2color)
cv2.imshow(title1, gray2gray)
cv2.imshow(title2, gray2color)
cv2.waitKey(0)  
```
