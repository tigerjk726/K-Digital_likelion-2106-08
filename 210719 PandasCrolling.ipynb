```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rc('font', family='NanumGothicCoding')
```

```python
plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.pointplot(x="연도 (년)", y="추계인구(명)", data=table[1])
a.grid()
plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.lineplot(x="연도 (년)", y="추계인구(명)", data=table[1])
a.grid()
```

```python
plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.pointplot(x="연도 (년)", y="출생자수(명)", data=table[1])
a.grid()
plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.lineplot(x="연도 (년)", y="출생자수(명)", data=table[1])
a.grid()
```

```python
plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.pointplot(x="연도 (년)", y="사망자수(명)", data=table[1])
a.grid()

plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.lineplot(x="연도 (년)", y="사망자수(명)", data=table[1])
a.grid()
```

```python
plt.figure(figsize = (18, 4))
plt.xticks(rotation=50)
a = sns.lineplot(x="연도 (년)", y="사망자수(명)", data=table[1])
a = sns.lineplot(x="연도 (년)", y="출생자수(명)", data=table[1])
a.grid()
```

```python
df = table[1]
df.columns
```

```python
df[['연도 (년)', '출생자수(명)','사망자수(명)']].plot().grid()
```

```python
df_pop = df[['연도 (년)', '출생자수(명)','사망자수(명)']]
df_pop = df_pop.set_index("연도 (년)")
df_pop.plot(figsize=(18,4)).grid()
```

```python
df_pop[-50:].plot(figsize=(18,4)).grid()
```

```python
plt.figure(figsize = (18, 4))
plt.grid()
sns.lineplot(x="연도 (년)", y="출생자수(명)", data=table[1][-50:])
sns.lineplot(x="연도 (년)", y="사망자수(명)", data=table[1][-50:])
```

```python
plt.figure(figsize = (18, 4))
plt.grid()
plt.xticks(rotation=50)
sns.barplot(x="연도 (년)", y='추계인구(명)', data=df)
```

```python
plt.figure(figsize = (18, 4))
plt.grid()
plt.xticks(rotation=50)
sns.barplot(x="연도 (년)", y='출생자수(명)', data=df)
sns.barplot(x="연도 (년)", y='사망자수(명)', data=df, palette='Blues')
```
