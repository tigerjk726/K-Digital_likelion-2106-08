```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.plot(np.linspace(0.1, 2.0, 20), np.logspace(0.02, 2.0, 20))
```

```python
data = [[2,81], [4,93], [6,91], [8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
# y= ax + b
#그래프 
plt.figure(figsize = (8,5))
plt.scatter(x, y)
plt.show()
#리스트로 된 xy값 넘파이로 배열
x_data = np.array(x)
y_data = np.array(y)

a=0
b=0

#학습률정하기
lr = 0.03

#몇번 반복될지 설정
epochs = 2001

for i in range(epochs):
    y_pred = a * x_data + b 
    error = y_data - y_pred
    #오차함수를 a로 미분, b로 미분
    a_diff = 2*(1/len(x_data))*sum((a * x_data + b - y_data)* x_data)
    b_diff = 2*(1/len(x_data))*sum(a * x_data + b - y_data)
    #학습률을 곱해 기존의 a,b값 업데이트
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print("epoch =%.f, 기울기=%.04f, 절편=%.04f"%(i, a, b))
        
y_pred = a * x_data + b
plt.scatter(x,y)
plt.plot([min(x_data),max(x_data)], [min(y_pred),max(y_pred)])
plt.show()
```
